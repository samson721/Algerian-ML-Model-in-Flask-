## Data Cleaning

```python
df[df.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
df=df
```

```python
df.info()
```

```python
df.head()
```

```python
df['Region']=df['Region'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df[df.isnull().any(axis =1)]
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.iloc[[122]]
```

```python
df= df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

### Changing the required columns to int and float

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
obj = [feature for feature in df.columns if df[feature].dtype =='O']
```

```python
obj
```

```python
for i in obj:
    if i!= 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('Aglerian_Forest_fire_Cleaned_dataset.csv',index=False)
```

##  Exploratory Data Analysis

```python
df1 = pd.read_csv('Aglerian_Forest_fire_Cleaned_dataset.csv')
```

```python
df1.head()
```

```python
df1= df1.drop(['day','month','year'],axis=1)
```

```python
df1.head()
```

```python
df1['Classes'].value_counts()
```

```python
df1['Classes'] =np.where(df1['Classes'].str.contains('not fire'), 0,1)
```

```python
df1.head()
```

```python
df1['Classes'].value_counts()
```

```python
import seaborn as sns
```

```python
# plot density plots for all features
plt.style.use('seaborn')
df1.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
# Percentage for pie chart
percentage = df1['Classes'].value_counts(normalize=True)*100
```

```python
# ploting pie charts
classlabel = ['Not fire','Fire']
plt.pie(percentage,labels = classlabel,autopct='%1.1f%%')
plt.show()
```
